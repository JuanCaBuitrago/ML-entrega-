```python
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
df_cars=pd.read_csv(r"C:\Users\Usuario\Documents\Proyecto ML\src\USA_cars_datasets.csv")


# Seleccionar las columnas necesarias
price = df_cars['price']
year = df_cars['year']
mileage = df_cars['mileage']

# Limpiar datos nulos para year y price
price_year = pd.DataFrame({'price': price, 'year': year}).dropna()

# Calcular el coeficiente de Pearson para year y price
coef_year, p_value_year = pearsonr(price_year['price'], price_year['year'])

print(f"Coeficiente de Pearson (price vs year): {coef_year}")
print(f"Valor p (price vs year): {p_value_year}")

# Visualización de dispersión para price vs year
plt.figure(figsize=(10, 6))
sns.scatterplot(data=price_year, x='year', y='price', alpha=0.7)
plt.title("Relación entre Year y Price")
plt.xlabel("Year")
plt.ylabel("Price")
plt.grid(True)
plt.show()

# Limpiar datos nulos para mileage y price
price_mileage = pd.DataFrame({'price': price, 'mileage': mileage}).dropna()

# Calcular el coeficiente de Pearson para mileage y price
coef_mileage, p_value_mileage = pearsonr(price_mileage['price'], price_mileage['mileage'])

print(f"Coeficiente de Pearson (price vs mileage): {coef_mileage}")
print(f"Valor p (price vs mileage): {p_value_mileage}")

# Visualización de dispersión para price vs mileage
plt.figure(figsize=(10, 6))
sns.scatterplot(data=price_mileage, x='mileage', y='price', alpha=0.7)
plt.title("Relación entre Mileage y Price")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.grid(True)
plt.show()

# Exploración por segmentos: Por ejemplo, marcas
if 'brand' in df_cars.columns:
    plt.figure(figsize=(14, 8))
    sns.boxplot(data=df_cars, x='brand', y='price')
    plt.title("Distribución de Price por Marca")
    plt.xlabel("Brand")
    plt.ylabel("Price")
    plt.xticks(rotation=45)
    plt.grid(True, axis='y', linestyle='--', alpha=0.7)
    plt.show()
```
