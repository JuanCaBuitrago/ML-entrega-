## Codificaci√≥n one-hot

```python
columns_to_encode = ['brand', 'model', 'title_status', 'color', 'state']

df_encoded = pd.get_dummies(df_cars, columns=columns_to_encode, prefix=columns_to_encode)
```

```python
from sklearn.preprocessing import LabelEncoder

# encoding a la columna 'condition'
le = LabelEncoder()
df_encoded['condition_encoded'] = le.fit_transform(df_encoded['condition'])
print(df_encoded.head)
```

```python
import matplotlib.pyplot as plt

years = [2021, 2027]
values = [195.84, 302.47]

plt.figure(figsize=(10, 6))
plt.bar(years, values, color=['blue', 'green'])
plt.title('Mercado de Autom√≥viles en Estados Unidos')
plt.xlabel('A√±o')
plt.ylabel('Valor en Miles de Millones de D√≥lares')
plt.ylim(0, 350)

for i, v in enumerate(values):
    plt.text(years[i], v + 5, f'${v}B', ha='center')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv('USA_cars_datasets.csv')

# Crear el gr√°fico
plt.figure(figsize=(15, 8))
sns.boxplot(x='state', y='price', data=df)
plt.xticks(rotation=90)
plt.title('Distribuci√≥n de precios por estado')
plt.xlabel('Estado')
plt.ylabel('Precio (USD)')
plt.show()
```
